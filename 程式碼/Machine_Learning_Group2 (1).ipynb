# 去除離異值（第一份資料)

```python
Q1=data_without_companyid_and_growth.quantile(0.25)
Q3=data_without_companyid_and_growth.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

# 排除第一份資料離異值

```python
dataset_cleared_iqr = data_without_companyid_and_growth[~((data_without_companyid_and_growth < (Q1 - 1.5 * IQR)) |(data_without_companyid_and_growth > (Q3 + 1.5 * IQR))).any(axis=1)]
dataset_cleared_iqr
```

# 計算第二份資料Q1,Q3,IQR

```python
Q1=data5_new.quantile(0.25)
Q3=data5_new.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

# 排除第二份資料離異值

```python
dataset_cleared_iqr_new = data5_new[~((data5_new < (Q1 - 1.5 * IQR)) |(data5_new > (Q3 + 1.5 * IQR))).any(axis=1)]
dataset_cleared_iqr_new
```

# 去除離異值的相關係數圖表（第二份資料）
