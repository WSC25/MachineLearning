```python
# 讀取 CSV File
from sklearn.ensemble import RandomForestClassifier
import pandas as pd # 引用套件並縮寫為 pd
df = pd.read_csv('C:/Users/wei/Documents/碩士/碩一/機器學習/mldata2.csv')
df
```

```python
df.info()
```

```python
col = ['ROA', 'ROE', 'Operating gross profit margin','cash flow','Debt ratio','Total asset turnover','kMeans_labels','rate']
df = df[col]
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['ROA', 'ROE', 'Operating gross profit margin','cash flow','Debt ratio','Total asset turnover','kMeans_labels']]
#y = df['Growth_CLass']
y = df['rate']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
df.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth = 4, n_estimators = 10)
rfc.fit(X_train, y_train)
```

```python
print('The accuracy of Random Forest Classifier on testing set:', rfc.score(X_test, y_test))
```

```python
testResults = rfc.predict(X_test)#rfc模型代號
```

```python
# Extract single tree
estimator_limited = rfc.estimators_[9]

from sklearn.tree import export_graphviz
export_graphviz(estimator_limited, out_file='tree5.dot', feature_names = X_train.columns,
                class_names = [str(i) for i in y_test.unique()],
                rounded = True, proportion = False, precision = 4, filled = True)
```
